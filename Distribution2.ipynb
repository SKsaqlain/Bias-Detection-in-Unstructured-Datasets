#### CODE TO PLOT FREQUENCY DISTRIBUTION.

```python

import numpy as np
from numpy import histogram,linspace
import pandas as pd
import re
from functools import reduce

from bokeh.plotting import figure
from bokeh.io import show,output_file,output_notebook,push_notebook
from bokeh.models import ColumnDataSource,Panel
from bokeh.models.tools import HoverTool
from bokeh.models.glyphs import VBar
from bokeh.models.widgets import CheckboxGroup,Tabs
from bokeh.layouts import column,row,WidgetBox

from bokeh.application.handlers import FunctionHandler
from bokeh.application import Application



output_notebook()
new_data=''
filename=''
def plot_bar(doc):
    def make_dataset(options=None):
        if options==None:
            return ColumnDataSource(new_data)
        temp_data=pd.DataFrame(columns=['x','top'])
        
        for ele in options:
            temp_data=temp_data.append(new_data.where(new_data.x==ele).dropna())
        
        #print(len(temp_data.x))
        return ColumnDataSource(temp_data)
        
    def make_bar_plot(src,filename='filename'):
        print(len(src.data['x']))
        p=figure(x_range=src.data['x'],plot_height=400,plot_width=400,title=filename,x_axis_label=filename,y_axis_label='frequency')
        p=figure(x_range=src.data['x'],plot_height=400,plot_width=400,title=filename,x_axis_label=filename,y_axis_label='frequency')
        glyph=VBar(x="x",top="top",bottom=0,width=0.5,fill_color="navy")
        p.add_glyph(src,glyph)
        
        p.xaxis.major_label_orientation=1.5
        hover=HoverTool(tooltips=[('frequency','@top'),(filename,'@x')])
        #p=style(p)
        p.add_tools(hover)
        p.legend.click_policy='hide'
        return p
    
    def update(attr,old,new):
        options=[checkBox.labels[i] for i in checkBox.active]
        new_src=make_dataset(options)
        src.data.update(new_src.data)
        #src.data['x']=new_src.data['x']
        #src.data['top']=new_src.data['top']
        
        
    
    checkBox=CheckboxGroup(labels=[str(ele) for ele in new_data.x],active=[i for i in range(len(new_data.x))])
    checkBox.on_change('active',update)
    
#     print(checkBox.active)
    initial_options=[checkBox.labels[i] for i in checkBox.active]
    src =make_dataset(initial_options)
    
    
    p=make_bar_plot(src,filename)
    controls=WidgetBox(checkBox)
    layout=row(controls,p)
    tab=Panel(child=layout,title=filename)
    tabs=Tabs(tabs=[tab])
    doc.add_root(tabs)
    #show(p)

    
    #p=style(p)
    

def process_data(filename):
    #processing data and converting it into dataFrame
    f=open(filename,'r',encoding='utf-8')
    data=f.readlines()
    #print(data)

    data_processed=[]
    for ele in data:
        #spliting data on white space and removing leading and ending whitespace and newline.
        x=re.split('\s',ele.strip(" \n"))
        #typecasting string number  to interger.
        x[0]=int(x[0])
        #preprocess
        #joining words 
        if(len(x)>2):    
            temp=' '.join(x[1:])
            x[1]=temp
        data_processed.append(x)
    #sorting the data in descending order
    data_processed.sort(key=lambda x:-x[0])
    x_axis=[ele[1] for ele in data_processed]
    y_axis=[ele[0] for ele in data_processed]

    new_data=pd.DataFrame({'x':x_axis,
                        'top':y_axis,})
    f.close()
    return new_data

def plot_bar_chart(fname='filename'):
    global filename
    filename=fname
    global new_data
    new_data=process_data(filename)
    
    handler=FunctionHandler(plot_bar)
    app=Application(handler)
    show(app)


plot_bar_chart('data/religion.txt')
```
